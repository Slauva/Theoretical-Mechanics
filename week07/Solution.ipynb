```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(x, *y, label, color="m-"):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.plot(x, *y, color)
    ax.legend(label)
    ax.set_xlabel(r"$time$")
    plt.grid()
    plt.show()
```

```python
def x_move(t):
    return 300625000 * (1 - np.exp(t / 9250)) + 32500 * t

def xi_move(t):
    return 42781250 * np.sqrt(3) * (np.exp(t / 9250) - 1) - 2.5 * np.sqrt(3) * t ** 2 + (3 - 4625 * np.sqrt(3)) * t
```

```python
T = np.linspace(0, 1)
x = x_move(T)
xi = xi_move(T)
```

```python
plot(T, x, xi, label=[r"$x(t)$", r"$\xi(t)$"])
```
